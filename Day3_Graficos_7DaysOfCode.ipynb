```python
'''Tarefa:
1 - CRIAR NOVA COLUNA (UTILIZANDO LOCALIZAÇÃO) DE ACORDO COM A CLASSE GERAL NA CDU
000 a 099: Generalidades. Ciência e conhecimento.
100 a 199: Filosofia e psicologia.
200 a 299: Religião.
300 a 399: Ciências sociais.
400 a 499: Classe vaga. Provisoriamente não ocupada.
500 a 599: Matemática e ciências naturais.
600 a 699: Ciências aplicadas.
700 a 799: Belas artes.
800 a 899: Linguagem. Língua. Linguística.
900 a 999: Geografia. Biografia. História

2- EXCLUIR REGISTRO_SISTEMA (não está fazendo sentido na análise) e transformar em formato String coluna MATRICULA_OU_SIAPE

Para mapear os valores, você pode usar uma lógica condicional de acordo com a numeração e armazená-los em uma lista.'''
```

```python
#Passo 1 verificaçao dados coluna localizacao
agrupamento['localizacao'].describe()
```

```python
# Transformar coluna localizacao do dataframe agrupamento para float, em seguida, para int
agrupamento['localizacao'] = pd.to_numeric(agrupamento['localizacao'], errors='coerce').fillna(0).astype(int)

# Transformar coluna localizacao para valor numerico inteiro
agrupamento['localizacao'] = pd.to_numeric(agrupamento['localizacao'], errors='coerce').fillna(0).astype(int)

#Verificar quantos Nan coluna localizacao
agrupamento['localizacao'].isna().sum()

```

```python
# Criando um dicionário para armazenar as classificações
classificacoes = {
    "Generalidades. Ciência e conhecimento.": [],
    "Filosofia e psicologia.": [],
    "Religião.": [],
    "Ciências sociais.": [],
    "Classe vaga. Provisoriamente não ocupada.": [],
    "Matemática e ciências naturais.": [],
    "Ciências aplicadas.": [],
    "Belas artes.": [],
    "Linguagem. Língua. Linguística.": [],
    "Geografia. Biografia. História.": [],
    "Classificação desconhecida.": []
}

# Modificando a função de classificação para preencher as listas
def classificar_cdu(localizacao):
    if isinstance(localizacao, int):
        codigo = localizacao

        if 0 <= codigo < 100:
            classificacoes["Generalidades. Ciência e conhecimento."].append(localizacao)
            return "Generalidades. Ciência e conhecimento."
        elif 100 <= codigo < 200:
            classificacoes["Filosofia e psicologia."].append(localizacao)
            return "Filosofia e psicologia."
        elif 200 <= codigo < 300:
            classificacoes["Religião."].append(localizacao)
            return "Religião."
        elif 300 <= codigo < 400:
            classificacoes["Ciências sociais."].append(localizacao)
            return "Ciências sociais."
        elif 400 <= codigo < 500:
            classificacoes["Classe vaga. Provisoriamente não ocupada."].append(localizacao)
            return "Classe vaga. Provisoriamente não ocupada."
        elif 500 <= codigo < 600:
            classificacoes["Matemática e ciências naturais."].append(localizacao)
            return "Matemática e ciências naturais."
        elif 600 <= codigo < 700:
            classificacoes["Ciências aplicadas."].append(localizacao)
            return "Ciências aplicadas."
        elif 700 <= codigo < 800:
            classificacoes["Belas artes."].append(localizacao)
            return "Belas artes."
        elif 800 <= codigo < 900:
            classificacoes["Linguagem. Língua. Linguística."].append(localizacao)
            return "Linguagem. Língua. Linguística."
        elif 900 <= codigo < 1000:
            classificacoes["Geografia. Biografia. História."].append(localizacao)
            return "Geografia. Biografia. História."
        else:
            classificacoes["Classificação desconhecida."].append(localizacao)
            return "Classificação desconhecida."
    else:
        classificacoes["Classificação desconhecida."].append(localizacao)
        return "Classificação desconhecida."
```

```python
#Aplicamdp a função para criar a nova coluna classificacao_cdu no dataframe agrupamento
agrupamento['classificacao_cdu'] = agrupamento['localizacao'].apply(classificar_cdu)
agrupamento
```

```python
#2- EXCLUIR REGISTRO_SISTEMA (não está fazendo sentido na análise) e transformar em formato String coluna MATRICULA_OU_SIAPE

# excluir coluna registro_sistema do dataframe
#agrupamento = agrupamento.drop(columns=['registro_sistema'])

# Transformar coluna matricula_ou_siape em string
agrupamento['matricula_ou_siape'] = agrupamento['matricula_ou_siape'].astype(str)
# Tirar esse .0 no final
agrupamento['matricula_ou_siape'] = agrupamento['matricula_ou_siape'].str.replace('.0', '', regex=False)

agrupamento.head()
```

```python
# 1. Análise Gráfica de Empréstimos:

# Converta as colunas do dataframe agrupamento de data para o tipo datetime
agrupamento['data_emprestimo'] = pd.to_datetime(agrupamento['data_emprestimo'])
agrupamento['data_devolucao'] = pd.to_datetime(agrupamento['data_devolucao'])

# Calcular a quantidade total de exemplares emprestados por cada ano
emprestimos_por_ano = agrupamento.groupby(agrupamento['data_emprestimo'].dt.year)['id_exemplar'].sum()

# Alterar a escala da coluna y do gráfico para um número inteiro
emprestimos_por_ano = emprestimos_por_ano.astype(int)

# Aumentar o tamanho da figura para melhorar a visualização
plt.figure(figsize=(10, 6))

# Ajustar a escala e remover a notação científica do eixo y do gráfico
emprestimos_por_ano.plot(kind='line', marker='o', title='Evolução dos Empréstimos Anuais')  # Adiciona marcadores nos pontos
plt.xlabel('Ano')
plt.ylabel('Quantidade de Exemplares Emprestados')

# Adicionar uma grade leve para facilitar a leitura dos valores
#plt.grid(True, linestyle='--', alpha=0.7)

# Ajustar o layout para evitar sobreposição de elementos
plt.tight_layout()

# Exibir o gráfico
plt.show()

# Criar um DataFrame a partir da série emprestimos_por_ano
tabela_emprestimos = emprestimos_por_ano.reset_index()
tabela_emprestimos.columns = ['Ano', 'Quantidade de Exemplares Emprestados']

# Filtrar do maior para o menor
tabela_emprestimos = tabela_emprestimos.sort_values(by='Quantidade de Exemplares Emprestados', ascending=False)

# Exibir a tabela de forma bonita com formatação brasileira
# Primeiro, converta a coluna para inteiro novamente, se necessário
tabela_emprestimos['Quantidade de Exemplares Emprestados'] = tabela_emprestimos['Quantidade de Exemplares Emprestados'].astype(int)

# Exibir a tabela de forma bonita
tabela_emprestimos.style.set_table_attributes('style="width: 50%; margin: auto; margin-top: 20px;"') \
                         .set_caption("Tabela de Empréstimos Anuais") \
                         .format({"Quantidade de Exemplares Emprestados": "{:,.0f}"})
```

```python
# 2. Análise Mensal dos Empréstimos

# Criar uma tabela com a quantidade total de exemplares emprestados por mês
emprestimos_por_mes = agrupamento.groupby(agrupamento['data_emprestimo'].dt.month)['id_exemplar'].sum()

# Formatar escala gráfico eixo x para nome do mês
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
emprestimos_por_mes.index = meses

# Aumentar o tamanho da figura para melhorar a legibilidade
plt.figure(figsize=(12, 6))

# Ajustar a escala no gráfico do eixo x para que permita ver todos os meses
emprestimos_por_mes.plot(kind='line', color='skyblue', title='Evolução dos Empréstimos Mensais')

# Rotacionar rótulos eixo x
plt.xticks(rotation=45)

plt.xlabel('Mês')
plt.ylabel('Quantidade de Exemplares Emprestados')
plt.tight_layout()  # Ajusta automaticamente o layout para que os rótulos não se sobreponham

# Adicionar marcadores nos pontos
plt.plot(emprestimos_por_mes.index, emprestimos_por_mes.values, marker='o')

plt.show()

# Analisar quais meses têm a maior quantidade de empréstimos, visualizando em tabela de ordem dos meses de maior valor para os menores
tabela_emprestimos_por_mes = emprestimos_por_mes.sort_values(ascending=False).reset_index()
tabela_emprestimos_por_mes.columns = ['Mês', 'Quantidade de Exemplares Emprestados']

# Exibir a tabela de forma bonita
tabela_emprestimos_por_mes.style.set_table_attributes('style="width: 50%; margin: auto; margin-top: 20px;"') \
                                 .set_caption("Tabela de Empréstimos Mensais") \
                                 .format({"Quantidade de Exemplares Emprestados": "{:,.0f}"})
```

**  *
Análise:
Março, Agosto, Fevereiro e Setembro são os meses de maior emprésticos. Provavelmente pelas semana de provas/ recuperações.**

```python
# 3. Análise Horária de Empréstimos:

#Verificar quais foram os horários com maior quantidade de empréstimos ao longo de um dia.
# Agrupar os dados por hora do dia e contar a quantidade de empréstimos em cada hora
emprestimos_por_hora = agrupamento.groupby(agrupamento['data_emprestimo'].dt.hour)['id_exemplar'].count()

# Plotar um gráfico de barras para visualizar os horários de maior movimento
plt.figure(figsize=(10, 6))  # Aumentar o tamanho da figura
emprestimos_por_hora.plot(kind='bar', color='skyblue', title='Empréstimos por Hora do Dia')  # Usar uma cor mais clara
plt.xlabel('Hora do Dia')
plt.ylabel('Quantidade de Empréstimos')
plt.grid(True, linestyle='--', alpha=0.5)  # Adiciona uma grade leve para melhor visualização
plt.tight_layout()  # Ajusta o layout para evitar sobreposição
plt.show()
#Inserir um espaçamento entre os gráficos
plt.subplots_adjust(hspace=3)


# Analisar os melhores horários para alocar atividades que não sejam de atendimento ao usuário com base nos dados coletados
# Analisar quais horários têm a menor quantidade de empréstimos, visualizando em tabela de ordem dos meses de maior valor para os menores
emprestimos_por_hora = emprestimos_por_hora.sort_values(ascending=True)

plt.figure(figsize=(10, 6))  # Aumentar o tamanho da figura
emprestimos_por_hora.plot(kind='bar', color='lightblue', title='Horários com Menor Movimento')  # Cor diferente para destacar
plt.xlabel('Hora do Dia')
plt.ylabel('Quantidade de Empréstimos')
plt.grid(True, linestyle='--', alpha=0.7)  # Adiciona uma grade leve para melhor visualização
plt.tight_layout()  # Ajusta o layout para evitar sobreposição
plt.show()
```
