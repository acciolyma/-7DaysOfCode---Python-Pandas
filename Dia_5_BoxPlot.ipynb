Fazendo Criando Função Tabela Frequência (mesma fórmula para geral)



Entendimento do Público da Biblioteca

```python
# Para incluir tanto a quantidade quanto o percentual na função tabela_frequencia, você pode modificar a função para que ela retorne um DataFrame com essas duas colunas.

def tabela_frequencia(coluna):
  quantidade = coluna.value_counts()
  percentual = coluna.value_counts(normalize=True) * 100
  tabela_frequencia = pd.DataFrame({'Quantidade':quantidade,'Percentual (%)':percentual.round(2)})
  return tabela_frequencia.round(2)

tabela_frequencia_vinculo = tabela_frequencia(emprestimos_completo['tipo_vinculo_usuario'])
tabela_frequencia_vinculo

```

Função gera_box_plot

```python
def gera_box_plot(dataset,x,y,titulo,subtitulo):
  '''
  Esta função irá gerar um gráfico de boxplot.

  Dataset = conjunto de dados do gráfico
  x = valor do eixo x do gráfico
  y = valor do eixo y do gráfico
  título = título do gráfico
  subtitulo = subtitulo do texto
  '''

  sns.set_theme(style="darkgrid", palette='Blues',font_scale=1.3)
  plt.figure(figsize=(16,10))

  ax = sns.boxplot(y= y, x= x, data= dataset,color='#4171EF')
  ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))


  plt.ylim(0,max(dataset[y])*1.1)                                               #Definir o limite do eixo y
  plt.xlabel(None)
  plt.ylabel(None)

  ax.set_title(titulo+"\n",size=20,loc='left',weight='bold')
  ax.text(s=subtitulo,x=-0.5,y=max(dataset[y])*1.11,fontsize=18, ha='left',color='gray')
```
